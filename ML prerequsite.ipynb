Question1:

The multiplication of (n,k) and (k,m) matrix is a (n,m) matrix. So, the each element of the resultant matrix has k multiplications and (k-1) additions. 
So, total no. of multiplications are n*m*k and additions are n*m*(k-1)

Question2:

```python
import numpy as np
import timeit
A = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
B = [[9, 8, 7], [6, 5, 4], [3, 2, 1]]
result = [[0,0,0],
         [0,0,0],
         [0,0,0]]
def matrix_list(A,B):
    for i in range(len(A)):
        for j in range(len(B[0])):
            for k in range(len(B)):
                result[i][j] += A[i][k] * B[k][j]
    return result
def matrix_numpy(A,B):
    mat = np.dot(A,B)
    return mat


list_time = timeit.timeit(lambda: matrix_list(A,B), number=100000)
numpy_time = timeit.timeit(lambda: matrix_numpy(A,B), number=100000)

print('Time for matrix multiplication using List:',list_time)
print('Time for matrix multiplication using numpy:',numpy_time)

#Because NumPy arrays are kept in memory continuously, as opposed to lists, processes can access and work with them with great efficiency.
    
```

Question5:

```python
import jax
import jax.numpy as jnp
import random
def f(x, y):
    return x**2 * y + y**3 * jnp.sin(x)
    
grad_fx = jax.grad(f, argnums=0)
grad_fy = jax.grad(f, argnums=1)

# randx = jnp.random.rand()
# randy = jnp.random.rand()
rand_x = 3.0
rand_y = 4.0
analy_grad_x = 2 * rand_x * rand_y + rand_y**3 * jnp.cos(rand_x)
analy_grad_y = rand_x**2 + 3 * rand_y**2 * jnp.sin(rand_x)

jax_grad_x = grad_fx(rand_x, rand_y)
jax_grad_y = grad_fy(rand_x, rand_y)

print("Gradients match with respect to x:", jnp.allclose(analy_grad_x, jax_grad_x))
print("Gradients match with respect to y:", jnp.allclose(analy_grad_y, jax_grad_y))
```

```python
import sympy as sp
x, y = sp.symbols('x y')
f = x**2 * y + y**3 * sp.sin(x)
grad_x = sp.diff(f, x)
grad_y = sp.diff(f, y)
print("Analytical Gradient with respect to x:", grad_x)
print("Analytical Gradient with respect to y:", grad_y)
```
